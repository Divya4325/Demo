```python
# Password Setting with atleat 8 characters
special_symbols="!@#$%^&*"
while True:
  password=input("Enter the Password:")
  if len(password)<8:
    print("Password must contain atleast 8 characters")
  elif not any(char.isdigit() for char in password):
    print("Password must contain atleast one digit")
  elif not any(char.isalpha() for char in password):
    print("Password must contain atleast one alpha character")
  elif not any(char in special_symbols for char in password):
    print("Password must contain atleast one special symbol")
  else:
    print("Password created Successfully")
    break
```
